```python
# Count and summarize the GTSRB training and test datasets using dataset_counter.py

import sys
import os

# Get the current working directory and navigate to dataset
current_dir = os.getcwd()
print(f"üìÇ Current directory: {current_dir}\n")

# Add current directory to path for imports
if current_dir not in sys.path:
    sys.path.append(current_dir)

from dataset_counter import CountDataset

# Construct absolute paths to train and test datasets
# Assuming you're running from src/utils/
workspace_root = os.path.abspath(os.path.join(current_dir, "../.."))
train_dataset_path = os.path.join(workspace_root, "dataset", "Train")
test_dataset_path = os.path.join(workspace_root, "dataset", "Test")

print(f"üîç Looking for Train dataset at: {train_dataset_path}")
print(f"üîç Looking for Test dataset at: {test_dataset_path}\n")

print("üöÇ TRAINING DATASET ANALYSIS")
print("=" * 80)
train_info = CountDataset(train_dataset_path)

print("\n\nüß™ TEST DATASET ANALYSIS")
print("=" * 80)
test_info = CountDataset(test_dataset_path)

# Optional: Compare train vs test distribution
if train_info and test_info:
    print("\n\nüìä TRAIN vs TEST COMPARISON")
    print("=" * 80)
    print(f"{'Metric':<30} {'Train':>15} {'Test':>15} {'Ratio':>15}")
    print("=" * 80)
    print(f"{'Total Samples':<30} {train_info['total_samples']:>15,} {test_info['total_samples']:>15,} {train_info['total_samples']/test_info['total_samples']:>14.2f}x")
    print(f"{'Number of Classes':<30} {train_info['num_classes']:>15} {test_info['num_classes']:>15} {'-':>15}")
    print(f"{'Total Size (MB)':<30} {train_info['total_size_mb']:>15.2f} {test_info['total_size_mb']:>15.2f} {train_info['total_size_mb']/test_info['total_size_mb']:>14.2f}x")
    print(f"{'Max Samples/Class':<30} {train_info['max_samples_per_class']:>15,} {test_info['max_samples_per_class']:>15,} {'-':>15}")
    print("=" * 80)
```
